## Get tweets that mention brands/products

Now we will make a simple query that retrieves all the the tweets that mention the brand or its products/players. We must consider the different names that each query could have. We oonly want the tweets that are in english, excluding also replies, retweeets and quotes.

```python
# Queries to execute (tweets)
# Videogames
q1 = 'nintendo lang:en -is:retweet -is:reply -is:quote'
q2 = '((fe engage) OR (fire emblem engage)) lang:en -is:retweet -is:reply -is:quote'
q3 = 'xbox lang:en -is:retweet -is:reply -is:quote'
q4 = 'hi-fi rush lang:en -is:retweet -is:reply -is:quote'
q5 = 'playstation lang:en -is:retweet -is:reply -is:quote'
q6 = 'forspoken lang:en -is:retweet -is:reply -is:quote'
# Football
q7 = 'chelsea lang:en -is:retweet -is:reply -is:quote'
q8 = '((manchester city) OR (cityzens)) lang:en -is:retweet -is:reply -is:quote'
q9 = 'liverpool lang:en -is:retweet -is:reply -is:quote' # No transfer in last day
q10 = '((manchester united) OR (red devils)) lang:en -is:retweet -is:reply -is:quote'
q11 = '((arsenal) OR (gunners)) lang:en -is:retweet -is:reply -is:quote'
q12 = '((tottenham hotspur) OR (spurs)) lang:en -is:retweet -is:reply -is:quote -(san antonio) -basketball -NBA'
q13 = 'cancelo lang:en -is:retweet -is:reply -is:quote'
q14 = 'sabitzer lang:en -is:retweet -is:reply -is:quote'
q15 = 'porro lang:en -is:retweet -is:reply -is:quote'
q16 = 'jorginho lang:en -is:retweet -is:reply -is:quote'
q17 = 'enzo fernandez lang:en -is:retweet -is:reply -is:quote'
```

```python
prueba = client.search_recent_tweets(query=q17, end_time="2023-02-07T23:00:00Z", tweet_fields=tweetFields, user_fields=userFields, place_fields=placeFields, media_fields=mediaFields, next_token=None, max_results=100)
for tweet in prueba['data']:
    print(tweet['text'])
```

```python
# Retrieval of tweets
pagTimes = 500
datos = client.search_recent_tweets(query=q17, end_time="2023-02-07T23:00:00Z", tweet_fields=tweetFields, user_fields=userFields, place_fields=placeFields, media_fields=mediaFields, next_token=None, max_results=maxResults)
for i in range(pagTimes-1):
    tweets = client.search_recent_tweets(query=q17, end_time="2023-02-07T23:00:00Z", tweet_fields=tweetFields, user_fields=userFields, place_fields=placeFields, media_fields=mediaFields, next_token=datos['meta']['next_token'], max_results=maxResults)
    # Add data from tweets that have been just read
    if ('data' in tweets): # Check whether there are more tweets to retrieve
        for tweet in tweets['data']:
            datos['data'].append(tweet)
    else:
        break
    # Update metadata
    datos['meta']['oldest_id'] = tweets['meta']['oldest_id']
    datos['meta']['result_count'] += tweets['meta']['result_count']
    if ('next_token' in tweets['meta']): # Check whether there are more tweets to retrieve
        datos['meta']['next_token'] = tweets['meta']['next_token']
    else:
        break
    print(datos['meta']['next_token'])
    # Avoid to exceed rate limit
    #time.sleep(1)
#for tweet in tweets['data']:
    #print(tweet['text'])
#print(tweets)
```

```python
# Save data into a json
with open('EnzoF.json', 'w') as fp:
    json.dump(datos, fp)
```
